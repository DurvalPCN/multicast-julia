<i>Abaixo implementação do <b>B-Multicast Assíncrono</b>, sendo executado com 10 processos adicionados.</i>

```julia
# Defining Assynchronous B-Multicast
function B_multicastAssync(g,m,B_deliver)
    for p in g
        remotecall(p, B_deliver, m)
    end
end

# Adding 10 processes (workers)
addprocs(10)

# My group is composed by all workers
g = workers()

# Multicasting "oi!" message to group `g` by assuming `println` as B_deliver function.
B_multicastAssync(g,"oi!",println)
```

Modifique a implementação do B-multicast para utilizar comunicação síncrona.

<i>Abaixo implementação do <b>B-Multicast Síncrono</b>, sendo executado com 10 processos adicionados.</i>

```julia
# Defining Synchronous B-Multicast
function B_multicastSync(g,m,B_deliver)
    for p in g
        remotecall_wait(p, B_deliver, m)
    end
end

# Multicasting "oi!" message to group `g` by assuming `println` as B_deliver function.
B_multicastSync(g,"oi!",println)
```

Execute as duas versões do B-multicast (síncrona e assíncrona) e compare as saídas.

<i>As duas versões, foram implementadas e executadas acima, bem como suas saídas.</i>

A ordem das mensagens das saídas foram as mesmas para ambas versões do B-multicast? Por quê?

**Resposta**:<br>
<i>Não.<br>
Na execução síncrona, há uma sincronização de relógio entre o receptor e emissor, fazendo com que a transmissão de dados ocorra seguindo a ordem correta de envio, no recebimento - ao contrário da execução assíncrona, que não existe sincronização de relógio parra distribuir a maneira de execução.</i>

### Exercício 3: Análise de desempenho do R-multicast (3,0 pontos)

Implemente duas versões do algoritmo R-multicast: uma assíncrona e outra síncrona.

* R-multicast assíncrono


```julia
# Flag to check if is received
isReceived = Int[]

# Defining function for Assynchronous R-Multicast
function R_multicastAssync(g, m, R_deliver)
    for p in g
        B_multicastRcvCheckAssync(isReceived, p, g, m, R_deliver)
    end
end

# Defining function for an Assynchronous B-Multicast that works checking flag isReceived from 
# Assynchronous R-Multicast, to check and communicate with all of the instanced elements from group
function B_multicastRcvCheckAssync(isReceived, p, g, m, R_deliver)
    if !(p in isReceived)
        push!(isReceived, p)
        remotecall(p, R_deliver, m)
        R_multicastAssync(g, m, R_deliver)
    end
end

R_multicastAssync(g, "hello!", println)
```

* R-multicast síncrono

```julia
# Flag to check if is received
isReceived = Int[]

# Defining function for Synchronous R-Multicast
function R_multicastSync(g, m, R_deliver)
    for p in g
        B_multicastRcvCheckSync(isReceived, p, g, m, R_deliver)
    end
end

# Defining function for a Synchronous B-Multicast that works checking flag isReceived from
# Synchronous R-Multicast, to check and communicate with all of the instanced elements from group
function B_multicastRcvCheckSync(isReceived, p, g, m, R_deliver)
    if !(p in isReceived)
        push!(isReceived, p)
        remotecall_wait(p, R_deliver, m)
        R_multicastSync(g, m, R_deliver)
    end
end

R_multicastSync(g, "hello!", println)
```
